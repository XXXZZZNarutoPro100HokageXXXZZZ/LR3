## Задание 1

Напишите код который стоит график функции y=x^2 на интервале [-10,10].
Добавьте заголовок подписей осей и сетку


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 11, 1)
y = x ** 2
plt.plot(x, y, label='y = x²', color='blue', linewidth=2)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.legend(fontsize=12)
plt.show()
```

## Задание 2

Постройте три линии на одном графике 

y = x

y = x^2

y = x^3

Добавьте легенду и сделайте оси одинакового масштаба

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = x
y2 = x**2
y3 = x**3

plt.plot(x, y, label='y = x', color='blue', linestyle='--', linewidth=2)  # Синяя пунктирная линия
plt.plot(x, y2, label='y = x^2', color='green', linestyle='-.', linewidth=2)  # Зеленая штрих-пунктирная линия
plt.plot(x, y3, label='y = x^3', color='red', linestyle='-', linewidth=2)  # Красная сплошная линия

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.legend(fontsize=12)
plt.axis('equal')  # Делает масштаб осей одинаковым
plt.show()
```

## Задание 7

Используя mpl_toolkits.mplot3d , постройте 3D-график функции z = sin((x^2 + y^2)^1/2)
на сетке значений x, y в диапазоне [-5,5]

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)  

Z = np.sin(np.sqrt(X**2 + Y**2))  # Правильная формула

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d') 

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('Ось X', fontsize=12)
ax.set_ylabel('Ось Y', fontsize=12)
ax.set_zlabel('Ось Z', fontsize=12)

plt.tight_layout()
plt.show()
```

## Задание 8

Постройте четыре графика в одной фигуре (2x2 сетка):

Линейный график y=x

Парабола y = x^2

Синус y = sin(x)

Косинус y = cos(x)

Добавьте заголовки к каждому подграфику.

```python
x = np.linspace(-np.pi, np.pi, 100)  
y1 = x
y2 = x**2
y3 = np.sin(x)  
y4 = np.cos(x)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0, 0].plot(x, y1, color = 'blue', linewidth = 2)
axes[0, 0].set_title('Линейная функция y = x')

axes[0, 1].plot(x, y2, color='red', linewidth=2)
axes[0, 1].set_title('Парабола: y = x²')

axes[1, 0].plot(x, y3, color='green', linewidth=2)
axes[1, 0].set_title('Синусоида: y = sin(x)')

axes[1, 1].plot(x, y4, color='brown', linewidth=2)
axes[1, 1].set_title('Косинусоида: y = cos(x)')


plt.show()
```

## Индивидуальное задание 3. Вариант 1

Вычислите определенный интеграл для функции:
f(x) = x^2
на отрезке [0, 3]
Постройте график и закрасьте область под кривой

```python
import numpy as np
from scipy.integrate import quad
def f(x):
    return x**2
    
x = np.linspace(-1, 2, 100)
y = f(x)

plt.plot(x, y, label='y=x^2', color='yellow', linewidth=2)
plt.fill_between(x, y, color='green', alpha=0.2)

plt.title('График функции $f(x) = x^2 + x$\nи площадь под кривой на отрезке [0, 3]', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend(fontsize=12)

integral, error = quad(f, -1, 2)
plt.show()
print(f"Значение интеграла:{integral}")
```

## Индивидуальное задание 4. Вариант 1

f(x, y) = x^2+y^2

Постройте 3D-график функции на интервале x, y ∈ [−5, 5].

```python

from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

Z = X**2+Y**2
fig = plt.figure(figsize=(10, 12))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('График функции f(x, y)= x²+y²', fontsize=14, pad=20)
ax.set_xlabel('Ось X', fontsize=12)
ax.set_ylabel('Ось Y', fontsize=12)
ax.set_zlabel('f(x, y)', fontsize=12)
plt.tight_layout()
plt.show()
```
